Q3. Write a program to perform arithmetic operations (Addition, Subtraction, Multiplication, Division).

```python
num1 = 20
num2 = 5

print("Addition =", num1 + num2)
print("Subtraction =", num1 - num2)
print("Multiplication =", num1 * num2)
print("Division =", num1 / num2)
```
