Q1. Create a list named shopping_list containing at least 5 initial items.

```python
shopping_list = ["milk", "bread", "eggs", "apples", "chicken"]
```

Q2. Print the entire list and also print the third item in the list.

```python
print("Shopping List =", shopping_list)

print("Third Item =", shopping_list[2])
```

Q3. Add two new items to the list using an appropriate list method.

```python
shopping_list.append("rice")
shopping_list.append("juice")

print("After Adding Items =", shopping_list)
```

Q4. Remove one item from the list.

```python
shopping_list.remove("bread")

print("After Removing Item =", shopping_list)
```

Q5. Create a dictionary named item_prices (at least 4 items with float prices).

```python
item_prices = {
    "milk": 2.5,
    "eggs": 3.0,
    "apples": 4.2,
    "chicken": 8.5
}
```

Q6. Print the price of one specific item from the dictionary.

```python
print("Price of Apples =", item_prices["apples"])
```

Q7. Print a formatted summary of your final shopping list and the price dictionary.

```python
print("\nFinal Shopping List:")
for item in shopping_list:
    print("-", item)

print("\nItem Prices:")
for item, price in item_prices.items():
    print(item, "=", price)
```
