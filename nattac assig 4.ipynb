```python
# Step 1 & 2: FizzBuzz using for loop (1 to 20)

for num in range(1, 21):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)


# Step 3: User se number lena
number = int(input("\nEnter a number for countdown: "))


# Step 5: Agar number 10 se zyada ho to break use karo
if number > 10:
    print("Number is greater than 10, stopping program!")
else:
    # Step 4: While loop se countdown
    while number >= 1:
        print(number)
        number -= 1

    print("Blast Off!")
```
