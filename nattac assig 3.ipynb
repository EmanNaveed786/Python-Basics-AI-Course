```python
# Personal Assistant Program

# Step 1: User se name lena
name = input("What is your name? ")

# Step 2: Birth year lena (integer me convert)
birth_year = int(input("What year were you born? "))

# Step 3: Current age lena (integer)
age = int(input("What is your current age? "))

# Step 4: Current year calculate karna
current_year = 2026   

# Calculate year when user turns 100
year_100 = current_year + (100 - age)

# Step 5: Lucky number lena (float me convert)
lucky_number = float(input("Enter your lucky number: "))

# Step 6: Summary print karna
print("\n--- Summary ---")
print("Name:", name)
print("Birth Year:", birth_year)
print("Current Age:", age)
print("You will turn 100 in the year:", year_100)
print("Your Lucky Number is:", lucky_number)
```
